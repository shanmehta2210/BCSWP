**Problem Solving**: Assignment 1: prints first 50 fibonacci numbers, and then from 1 to 50 types fizz if multiple of 3, buzz if multiple of 5, fizzbuzz if both.

```python
a = 1
b = 1
print(a, end=" ")
for i in range(49):
  print(b, end=" ")
  b = a + b
  a = b - a
print("\n")
for i in range(1, 51):
  if i%3 == 0 and i%5 == 0:
    print("FizzBuzz", end=" ")
    continue
  if i%3 == 0:
    print("Fizz", end=" ")
  if i%5 == 0:
    print("Buzz", end=" ")
```
